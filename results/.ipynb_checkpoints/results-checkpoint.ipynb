# Plot results for each active learning strategy, dataset, language model, information retrieval metric and machine learning model 

Follow these steps:
- Choose the active learning strategy and instantiate the **al** variable with the desired strategy to plot, which can be **UNCERTAINTY** (for uncertainty sampling active learning strategy) or **RANDOM** (for random sampling).  
- Choose the dataset and instantiate the **dataset** variable with the desired dataset to plot the results, which can be **clef** (for HealthCLEF dataset) or **episte** (for Epistemonikos dataset). 
- Choose the language model and instantiate the **language_model** variable with the desired language model to plot, the alteratives are **glove**, **w2vec**, **BERT**, and **BioBERT**.   
- Machine Learning models are always compared to each other.  
- Finally, choose the information retrieval metrics to compare, they can be recall@10 (r@10), recall@20 (r@20), recall@30 (**r@30**), precision@10 (**p@10**), precision@20 (**p@20**), precision@30 (**p@30**), MAP (**MAP**, for mean average precision), lastrel (**lastrel**, for last relevant found) and WSS (**WSS**, for work saved oversampling). 

Then with this setup, results can be plotted and compared considering results and their standard deviation (error bar).

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import xticks
from matplotlib.pyplot import ylim
%matplotlib inline
```

```python
al = 'UNCERTAINTY' # choose UNCERTAINTY or RANDOM active learning strategy 
dataset = 'episte' # choose clef or episte 
language_model = 'w2vec' # choose glove, w2vec, BERT, BioBERT
models = ['MLP', 'SVM', 'LR', 'RF']
metric = 're@10' # choose r@10, r@20, r@30, p@10, p@20, p@30 , MAP, lastrel, WSS

COLUMNS = [al + dataset + language_model + m + metric for m in models ]

# results 
df = pd.read_csv('results/{}_results.csv'.format(dataset))

# standard deviation 
df_stdevs = pd.read_csv('results/{}_results_stdevs.csv'.format(dataset))

df[COLUMNS]
```

```python
df_results = df[COLUMNS]
df_stdevs_results = df_stdevs[COLUMNS]

#change columns names 
df_results.columns = [[ language_model + m for m in models ]]

df_results['x'] = range(1,11)

df_results
```

## Plot

```python
plt.errorbar(
    [i for i in range(10)],  # X
    [x[0] for x in df_results['{}{}'.format(language_model, models[0])].values.tolist()], # Y
    yerr=sum(df_stdevs_results[al + dataset + language_model + models[0] + metric])/10,     # Y-average-errors
    fmt="o", # format line like for plot()
    linewidth=1,# width of plot line
    markersize=8,
    color = 'blue',
    capsize=5,
    ecolor='blue',
    label='{}{}'.format(language_model,models[0]),
    linestyle='dashed'
    )

plt.errorbar(
    [i for i in range(10)],  # X
    [x[0] for x in df_results['{}{}'.format(language_model, models[1])].values.tolist()], # Y
    yerr=sum(df_stdevs_results[al + dataset + language_model + models[1] + metric])/10,     # Y-errors
    fmt="o", # format line like for plot()
    linewidth=1,# width of plot line
    markersize=8,
    color = 'green',
    capsize=5,
    ecolor='green',
    label='{}{}'.format(language_model,models[1]),
    linestyle='dashed'
)


plt.errorbar(
    [i for i in range(10)],  # X
    [x[0] for x in df_results['{}{}'.format(language_model, models[2])].values.tolist()], # Y
    yerr=sum(df_stdevs_results[al + dataset + language_model + models[2] + metric])/10,     # Y-errors
    fmt="o", # format line like for plot()
    linewidth=1,# width of plot line
    markersize=8,
    color = 'orange',
    capsize=5,
    ecolor='orange',
    label='{}{}'.format(language_model,models[2]),
    linestyle='dashed'
)

plt.errorbar(
    [i for i in range(10)],  # X
    [x[0] for x in df_results['{}{}'.format(language_model, models[3])].values.tolist()], # Y
    yerr=sum(df_stdevs_results[al + dataset + language_model + models[3] + metric])/10,     # Y-errors
    fmt="o", # format line like for plot()
    linewidth=1,# width of plot line
    markersize=8,
    color = 'red',
    capsize=5,
    ecolor='red',
    label='{}{}'.format(language_model,models[3]),
    linestyle='dashed'
)


# change ylim to 0, 0,4 for HealthCLEF
ylim(0,1)


plt.ylabel("recall@10")
plt.xlabel("iterations")
plt.legend(loc='lower right')
plt.title(' {} document representation {} Sampling'.format(language_model.capitalize(), al.capitalize()) )
```
