```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import xticks
from matplotlib.pyplot import ylim
%matplotlib inline
```

```python
df = pd.read_csv('results/episte_results.csv', sep=';')

df
```

```python
al = 'UNCERTAINTY' # choose UNCERTAINTY or RANDOM active learning strategy 
dataset = 'episte' # choose clef or episte 
language_model = 'BERT' # choose glove, w2vec, BERT, BioBERT
models = ['MLP', 'SVMrbf', 'LR', 'RF']
metric = 're@10' # choose r@10, r@20, r@30, p@10, p@20, p@30 , MAP, lastrel, WSS

COLUMNS = [al + dataset + language_model + m + metric for m in models ]


# standard deviation 
df_stdevs = pd.read_csv('results/{}_results_stdevs.csv'.format(dataset), sep=';' )

df[COLUMNS]
```

```python
df_stdevs
```

```python
df_results = df[COLUMNS]
df_stdevs_results = df_stdevs[COLUMNS]

#change columns names 
df_results.columns = [[ language_model + m for m in models ]]

df_results['x'] = range(1,11)

df_results
```

```python
# results relevance feeedback 
relevance_feedback = [0.07936865402105353,
 0.13064417401568865,
 0.16639958529413612,
 0.1880397972613835,
 0.19766267078056482,
 0.21003760493733672,
 0.21333560724059542,
 0.22111005113039342,
 0.2279362521761895,
 0.23009618488525538]
```

```python
plt.errorbar(
    [i for i in range(10)],  # X
    [x[0] for x in df_results['{}{}'.format(language_model, models[0])].values.tolist()], # Y
    yerr=sum(df_stdevs_results[al + dataset + language_model + models[0] + metric])/10,     # Y-average-errors
    fmt="o", # format line like for plot()
    linewidth=1,# width of plot line
    markersize=8,
    color = 'blue',
    capsize=5,
    ecolor='blue',
    label='{}{}'.format(language_model,models[0]),
    linestyle='dashed'
    )

plt.errorbar(
    [i for i in range(10)],  # X
    [x[0] for x in df_results['{}{}'.format(language_model, models[1])].values.tolist()], # Y
    yerr=sum(df_stdevs_results[al + dataset + language_model + models[1] + metric])/10,     # Y-errors
    fmt="o", # format line like for plot()
    linewidth=1,# width of plot line
    markersize=8,
    color = 'green',
    capsize=5,
    ecolor='green',
    label='{}{}'.format(language_model,models[1]),
    linestyle='dashed'
)


plt.errorbar(
    [i for i in range(10)],  # X
    [x[0] for x in df_results['{}{}'.format(language_model, models[2])].values.tolist()], # Y
    yerr=sum(df_stdevs_results[al + dataset + language_model + models[2] + metric])/10,     # Y-errors
    fmt="o", # format line like for plot()
    linewidth=1,# width of plot line
    markersize=8,
    color = 'orange',
    capsize=5,
    ecolor='orange',
    label='{}{}'.format(language_model,models[2]),
    linestyle='dashed'
)

plt.errorbar(
    [i for i in range(10)],  # X
    [x[0] for x in df_results['{}{}'.format(language_model, models[3])].values.tolist()], # Y
    yerr=sum(df_stdevs_results[al + dataset + language_model + models[3] + metric])/10,     # Y-errors
    fmt="o", # format line like for plot()
    linewidth=1,# width of plot line
    markersize=8,
    color = 'red',
    capsize=5,
    ecolor='red',
    label='{}{}'.format(language_model,models[3]),
    linestyle='dashed'
)



plt.errorbar(
     [i for i in range(10)],  # X
     relevance_feedback, # Y
     yerr=0.05,     # Y-errors
     fmt="o", # format line like for plot()
     linewidth=0.5,# width of plot line
     markersize=8,
     color = 'grey',
     capsize=5,
     ecolor='grey',
     label='r.feedback',
     linestyle='dashed'
 )


ylim(0,0.8)


plt.ylabel("recall@10")
plt.xlabel("iterations")
plt.legend(loc='lower right')
plt.title(' {} document representation {} Sampling'.format(language_model.capitalize(), al.capitalize()) )

# save plot
plt.savefig('{}_{}_{}.eps'.format(dataset, language_model, al ), format='eps')
```
